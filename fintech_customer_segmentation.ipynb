```python
import pandas as pd

df = pd.read_csv('/bank-additional-full.csv', sep=';')

df.head()


```

```python
# Basic structure of the dataset
df.info()
```

```python
# Missing values
df.isnull().sum()
```

```python
df.columns.tolist()
```

```python
import matplotlib.pyplot as plt

# Plot job type counts
df['job'].value_counts().head(10).plot(kind='bar', title='Top 10 Job Types')
plt.ylabel('Number of Customers')
plt.xlabel('Job Title')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['education'].value_counts().plot(kind='bar', title='Education Level Distribution')
plt.ylabel('Number of Customers')
plt.xlabel('Education Level')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create age groups
df['age_group'] = pd.cut(df['age'], bins=[17, 25, 35, 50, 65, 100],
                         labels=['18â€“25', '26â€“35', '36â€“50', '51â€“65', '66+'])

# Plot age group distribution
df['age_group'].value_counts().sort_index().plot(kind='bar', title='Customer Age Groups')
plt.ylabel('Number of Customers')
plt.xlabel('Age Group')
plt.grid(True)
plt.tight_layout()
plt.show()
```
